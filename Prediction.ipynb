```python

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
data_good=pd.read_csv('data_good.csv')
data=pd.read_csv('creditcard.csv', error_bad_lines=False)
data_prueba=pd.read_csv('prueba.csv') #set de testeo
Valores = pd.read_csv('Valores.csv')
Epsilon=pd.read_csv('Epsilon.csv')

epsilon=Epsilon.loc[0][0]/2
print(epsilon)


keys = ['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
y = ['Class']
u=Valores[keys].loc[0]
o=Valores[keys].loc[1]




```

```python
def gauss(x,u,o):
    distribution=(1/(o*(2*np.pi)**(1/2)))*np.exp(-(1/2)*((x-u)**2/(o**2)))

    return(distribution/1000.0)




def proba(x):
    
    prob=1
    
    for j in range(len(u)):
        distribution=gauss(x[j],u[j],o[j])
        prob=prob*distribution
    return(prob)
```

```python
def clasificador(x):
    p=proba(x)
    if p<epsilon:
        return int(1)
    if p>=epsilon:
        return int(0)
    else:
        return int(2)
    
```

```python
%%time
prediccion=[]
valor_proba=[]

for i in range(len(data[keys])):
    x=data[keys].loc[i].tolist()
    resultado=clasificador(x)
    prediccion.append(resultado)
    if(i%10000==0):
        print(i)
```
