```python
#calcular las gaussianas para cada uno

norm_g=[]
norm_g1=[]

def gauss_1(x,u,o):
    distribution=(1/(o*(2*np.pi)**(1/2)))*np.exp(-(1/2)*((x-u)**2/(o**2)))
    norm1=sum(distribution)
    distribution=distribution/norm1
    norm_g1.append(norm1)
    return distribution

def gauss(x,u,o):
    dist=0
    for i in range(len(u)):
        dist=dist+gauss_1(x,u[i],o[i])
    
    norm=sum(dist)
    dist=dist/norm
    norm_g.append(norm)
    return dist
```
